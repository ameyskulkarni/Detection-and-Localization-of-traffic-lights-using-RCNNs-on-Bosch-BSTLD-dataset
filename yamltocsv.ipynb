```python
import yaml
import csv
import json
```

```python
yaml_file= './train.yaml'
with open(yaml_file) as f:
        data = yaml.load(f)
```

```python
rows_to_write = []
```

```python
print(data[1]['boxes'])
print(len(data))
print(len(data[1]['boxes']))
print(type(data[2]['boxes'][0]['label']))
```

```python
for i in range(0,len(data)):
    if(len(data[i]['boxes'])==0):
        continue
    pth=data[i]['path']
    for j in range(0,len(data[i]['boxes'])):
        lab=data[i]['boxes'][j]
        xmin=data[i]['boxes'][j]['x_min']
        xmax=data[i]['boxes'][j]['x_max']
        ymin=data[i]['boxes'][j]['y_min']
        ymax=data[i]['boxes'][j]['y_max']
        label=data[i]['boxes'][j]['label']
        rows_to_write.append([pth,xmin,xmax,ymin,ymax,label])
        
with open('output_csv_file_new.csv', 'w') as out:
    csv_writer = csv.writer(out, delimiter=',')
    csv_writer.writerows(rows_to_write)
    print("Output file output_csv_file.csv created")
```
